```python
img_array = np.array(img, dtype=np.float)
mnist_array = np.array(mnist, dtype=np.float)
```

```python
h,w = img_array.shape
img_array = img_array.reshape((h,w,1))
```

```python
mnist_array.shape
```

```python
img_roll = np.roll(img_array, shift= 3, axis=1)
mnist_roll = np.roll(mnist_array, shift=100, axis=1)
```

```python
array_buffer = img_array.tobytes()
im = Image.new("I", img_array.T.shape)
im.frombytes(array_buffer, 'raw', "I;16")
im.save("16-bit_test_pillow.png")
```
